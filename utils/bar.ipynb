```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

labels = ['cpu+mem', 'cpu+disk', 'cpu+net', 'cpu+mem+disk', 'cpu+disk+net','cpu+mem+net','cpu+mem+disk+net'] # 
men_means = [0.0563,0.0441,0.0573,0.0462,0.0560,0.0591,0.0514]
women_means = [0.0796,0.0639,0.0809,0.0696,0.0815,0.0822,0.0770]
wm_means = [0.1097,0.0920,0.1115,0.0939,0.1162,0.1149,0.1025]

x = np.arange(len(men_means))

plt.figure(figsize=(27,12))

width = 0.95

rects1 = plt.bar(x - width/3, men_means, width/3) # 
rects2 = plt.bar(x, women_means, width/3)
rects3 = plt.bar(x + width/3, wm_means, width/3)

# 
plt.ylabel('Error', fontsize=25)
plt.yticks(fontsize=22)
# plt.title('Scores by group and gender')
plt.xticks(x,labels, rotation=5, fontsize=30)
plt.legend(['MAE','RMSE','MAPE'], fontsize=20)

# 
def set_label(rects):
    for rect in rects:
        height = rect.get_height() # 
        plt.text(x = rect.get_x() + rect.get_width()/2, # 
                 y = height + 0.005, # 
                 s = height, # 
                 ha = 'center', fontsize=15) # 

set_label(rects1)
set_label(rects2)
set_label(rects3)

# plt.tight_layout() # 
# plt.savefig('./.png')
```
